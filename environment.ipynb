```python
from environment import Environment
env = Environment("IMG_2684.MOV")
# env = Environment("shortened.mov")
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib qt

points = np.vstack(env.point_cloud)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = points[:, 0]
y = points[:, 1]
z = points[:, 2]

ax.scatter(x, y, z, c='b', marker='o', s=1)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
